```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
x =  [1, 2, 3, 4, 5]
y =  [10, 12, 8, 15, 7]
sns.lineplot(x =x, y=y)
```

```python
x =  [1, 2, 3, 4, 5]
y =  [10, 12, 8, 15, 7]
plt.plot(x,y, linestyle = '--', color = 'g', marker ='o')
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values = [25, 40, 15, 30]
plt.figure(figsize =(8,4))
plt.bar(categories,values,color = 'blue')
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values = [25, 40, 15, 30]

sns.barplot(x = categories, y = values, )
plt.show()
```

```python
data = [32, 45, 50, 28, 36, 42, 38, 25, 29, 52, 48, 60]

sns.histplot(x = data, bins =10, kde=True)
plt.show()
```

```python
data = [32, 45, 50, 28, 36, 42, 38, 25, 29, 52, 48, 60]
plt.hist(data, color = 'g', bins = 10)
plt.show()
```

```python
x = [3, 5, 7, 10, 12, 15]
y = [25, 40, 30, 50, 35, 60]

sns.scatterplot(x= x, y = y)
plt.show()
```
